```python
import numpy as np
```

```python
# Samples per ms
smplr = 2000
minPktLen = 100
```

```python
bin = np.fromfile('112-Char-10-SF.complexfloat32', dtype=np.complex64)
```

```python
plot(np.array(range(len(bin)))/2000, abs(bin))
```

```python
abin = abs(bin)
bbin = abin>0.1
plot(np.array(range(len(bin)))/2000, bbin)
```

```python
d = dict()
start = 0
for i in range(1,len(bbin)):
    if (not bbin[i-1]) and bbin[i]:
        start = i
    elif bbin[i-1] and (not bbin[i]):
        if(i-start > minPktLen):
            d[start/smplr] = (i-start)/smplr
print(d)
```
